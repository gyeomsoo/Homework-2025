# 178 향상된 에코 클라이언트 만들기 : socket

```python
import socket

HOST = 'localhost'
PORT = 7777
def echoclient2():
  with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
    sock.connect((HOST, PORT))

    while True:
      msg = input('메시지 입력: ')
      if msg == '/q':
        sock.sendall(msg.encode())
        break

      sock.sendall(msg.encode())
      data = sock.recv(1024)
      print(f'에코 서버로부터 받은 메시지 [{data.decode()}]')

    print('클라이언트 종료')

echoclient2()
```

# 179 웹사이트에 접속하여 HTML 페이지를 파일로 저장 : urllib.request.urlopen()

```python
form urllib.request import urlopen

url = 'https://www.python.org/'
with urlopen(url) as f:
  doc = f.read().decode()
  with open('pythonhome.html', 'w') as h:
    h.write(doc)
```
