# 186 비동기 함수 구현하기 : asyncio, async, await

```python
import asyncio
from time import time

async def sayAfter(delay, saying):
  print(f'sayAfter({delay}, saying) 시작...')
  await asyncio.sleep(delay)
  print(saying)
```

```python
async def main():
  print('비동기 함수 main() 시작...')
  stime = time()
  await sayAfter(1, '안녕하세요!')
  await sayAfter(2, '방가방가~~')
  etime = time()
  print(f'총 소요 시간: {etime-stime:.2f}초')
```

```python
async def main():
  print('비동기 함수 main() 시작...')
  tasks = [ ]
  tasks.append(asyncio.create_task(sayAfter(1, '안녕하세요!')))
  tasks.append(asyncio.create_task(sayAfter(2, '방가방가~~')))

  stime = time()
  await asyncio.wait(tasks)
  etime = time()
  print(f'총 소요 시간: {etime-stime:.2f}초')

await main()
```

# 187 opencv를 활용하여 마우스로 도형 그리기 1

```python
import cv2

def showImage(imgfile):
  img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
  cv2.namedwindow('picture', img)
  cv2.imshow('picture', img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

showImage('sunrise.jpg')
```

```python
import numpy as np
from random import shuffle

B = [i for i in range(256)]
G = [i for i in range(256)]
R = [i for i in range(256)]
```

```python
def onMouse(event, x, y, flags, param):
if event == cv2.EVENT_LBUTTONDBLCLK:
  shuffle(B), shuffle(G), shuffle(R)
  cv2.circle(param, (x, y), 50, (B[0], G[0], R[0]), -1)
```

```python
def mouseBrush():
  img = np.zeros((576, 1024, 3), np.uint8)
  cv2.namedwindow('drawing pad')
  cv2.wetMouseCallback('drawing pad', onMouse, img)

  while True:
    cv2.imshow('drawing pad', img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
      break
  cv2.destroyAllWindows()

mouseBrush()
```
