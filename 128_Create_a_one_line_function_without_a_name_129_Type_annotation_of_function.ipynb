# 128 이름없는 한 줄짜리 함수 만들기 : lambda

```python
lambda 매개변수1, 매개변수2, ...:리턴식
```

```python
f = lambda x: x**2
print(f(2))
print(f(3))
```

```python
f = lambda x: x + '.txt'
print(f('readme'))
```

```python
f = lambda x: x[1]
print(f('readme'))
print(f(('장비', 10)))
```
