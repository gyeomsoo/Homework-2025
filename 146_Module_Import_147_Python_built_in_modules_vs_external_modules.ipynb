# 146 모듈 임포트 : import, import ~ as, from ~ import

```python
import 모듈 이름
import 패키지 계층.모듈 이름
```

```python
import math
print(f'sqrt(2) = {math.sqrt(2):.3f}')
```

```python
import 모듈 as 별명
```

```python
import math as m
print(f'sqrt(2) = {m.sqrt(2):.3f}')
```

```python
from 모듈 import 함수1, 함수2...
from 패키지 import 모듈1, 모듈2...
from 패키지 계층 import 함수1, 함수2...
```

```python
from math import sqrt, log
print(f'sqrt(2) = {sqrt(2):.3f}')
print(f'log(2) = {log(2):.3f}')
```
