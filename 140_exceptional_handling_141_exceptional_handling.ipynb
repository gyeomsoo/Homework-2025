# 140 예외처리 4 : try ~ except Exception as e

```python
intList = range(10)
idx = 10
try:
  print(intList[idx])
except Exception as e:
  exceptionClass = e.__class__.__name__
  exceptionDetaill = e.__str__()                # print(e) = print(e._str_())
  print(f'예외종류: {exceptionClass}, 예외내용: {exceptionDetaill}')
```

# 141 예외처리 5 : try ~ except 특정 예외

```python
try:
  잠재적으로 오류 및 예외가 발생할 수 있는 코드
except: 예외클래스 1:
  예외 1이 발생했을 때 예외를 처리하는 코드
except: 예외클래스 2:
  예외 2가 발생했을 때 예외를 처리하는 코드
```

```python
intList = range(10)
try:
  idx = int(input("숫자를 입력하세요:"))
  print(intList[idx])
except ValueError:
  print('입력한 내용이 숫자가 아닙니다!')
except IndexError as e:
  print(f'인덱스 오류가 발생: {e}')
```
