# 190 matplotlib을 활용한 데이터 시각화 실습 2

```python
import matplotlib.pyplot as plt
import csv
```

```python
file = 'births.txt'
years = range(1800, 2012)
pieces = [ ]
with open(file, 'r') as f:
  data = csv.reader(f, delimiter=',')
  for d in data:
    pieces.append(d)
```

```python
x = [int(year) for year, female, male in pieces]
y1 = [int(female) for year, female, male in pieces]
y2 = [int(male) for year, female, male in pieces]

plt.plot(x, y1, 'r-', label='여자아기')
plt.plot(x, y2, 'b-', label='남자아기')

plt.ylim([0, 2500000])
curYticks, _= plt.yticks()
plt.yticks(curYticks, [f'{y:,.0f}' for y in curYticks])

plt.legend()
plt.title('미국 연도별 성별 총 출생아수')
plt.xlabel('연도')
plt.ylabel('출생아수')
plt.grid()
plt.show()
```

# 191 Basemap을 활용하여 지도 출력하기

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill')
m.drawcoastlines()
m.drawcountries()
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill')
m.drawcoastlines()
m.drawcountries()
m.bluemarble()
plt.show()
```

```python
import numpy as np

m = Basemap(projection='robin', resolution='l', lat_0=0, lon_0=128)
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray')
m.drawmapboundary()
m.drawmeridians(np.arange(0, 360, 30))
m.drawparallels(np.arange(-90, 90, 30))
plt.show()
```
