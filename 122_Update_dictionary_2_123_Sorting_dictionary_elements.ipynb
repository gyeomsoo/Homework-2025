# 122 딕셔너리 갱신하기 2 : *d1* | *d2*, *d1* | = *d2*

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
others = {'제갈량':99, '황충':70}
update = greats | others
print(update)
```

```python
updated2 = update | {'유비':90, '관우':96}
print(updated2)
```

```python
greats |= others        # greats.updete(others)와 동일
print(greats)
```

# 123 딕셔너리 요소 정렬하기 : sorted ()

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
ret = sorted(greats)
print(ret)
```

```python
ret = sorted(greats.items())
print(ret)
```

```python
ret = sorted(greats.items(), reverse=True)
print(ret)
```

```python
ret = sorted(greats.items(), key=lambda x: x[1])
print(ret)
```

```python
ret = sorted(greats.items(), key=lambda x: x[1], reverse=True)
print(ret)
```

```python
for name, level in ret:
  print(f'<{name}>의 레벨은 <{level}>입니다.')
```
