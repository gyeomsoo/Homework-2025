# 160 파일 이름 변경/파일 이동 : os.rename ()

```python
from os import rename

src = 'sunrise_copied.jpg'
dst = 'sunrise2.jpg'
rename(src, dst)
```

# 161 디렉토리에 존재하는 파일 목록 얻기 : os.listdir (), glob.glob ()

```python
from os import listdir

forder = './'             # 현재 작업 디렉토리
ret = listdir(forder)
print(ret)
```

```python
from glob import glob

target = '*.txt'
ret = glob(target)
print(ret)
```

```python
target = '**'
ret = glob(target, recursive = True)
ret.sort()
for r in ret:
  print(r)
```
