# 180 HTTP 헤더 조작하기 : requests

```python
import requests

def my_urlopen(url):
  header = {"User-Agent":"Mozilla/5.0 (windows NT 10.0; win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/97.04692.99 Safari/537.36"}
    try:
      with requests.get(url, headers=header) as r:
        return r
    except Exception as e:
      print(f'my_urlopen() 오류: url={url}\n{e}')
      return None
```
