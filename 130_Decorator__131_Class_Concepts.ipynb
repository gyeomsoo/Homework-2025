# 130 데코레이터 : @

```python
def report():
  print('이 부분은 리포트의 본문입니다.')
```

```python
def makeNumber():
  for i in range(3):
    print(i)
```

```python
def report1():
  print('-'*50)                           # 출력물의 위쪽 점선
  print('이 부분은 리포트의 본문입니다.')
  print('-'*50)                           # 출력물의 아래쪽 점선
  return

def makeNumber1():
  print('-'*50)                           # 출력물의 위쪽 점선
  for i in range(3):
    print(i)
  print('-'*50)                           # 출력물의 아래쪽 점선
  return
```

```python
def makeForm(func):
  def drawLine():
    print('-'*50)
    func()
    print('-'*50)
  return drawLine
```

```python
report_new = makeForm(report)
report_new()
```

```python
def drawLine():
  print('-'*50)
  report()
  print('-'*50)
```

```python
makeNumber_new = makeForm(makeNumber)
makeNumber_new()
```

```python
@makeForm                     # 이 부분이 데코레이터
def report():
  print('이 부분은 리포트의 본문입니다.')
  return
```

```python
# report()를 실행하면 아래의 코드와 같이 수행하세요
ret = makeForm(report)
ret()
```

```python
def decoFunc(func):
  def wrapper(*args, **kwargs):
    # 공통 기능 수행 코드
      func(*args, **kwargs)
    # 공통 기능 수행 코드
  return
```
