# 172 오늘의 요일 계산하기 : time.localtime()

```python
from time import localtime, strftime

weekdays = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
t = localtime()
today = strftime('%y-%m-%d', t)
print(f'{today}은 {weekdays[t.tm_wday]}입니다.')
```

# 173 정규표현식 개념

```python
ab?c : a와 c사이에 b가 0 또는 1회 나타나는 문자열과 매치됨
ab*c : a와 c사이에 b가 0회 이상 반복으로 나타나는 문자열과 매치됨
ab+c : a와 c사이에 b가 1회 이상 반복으로 나타나는 문자열과 매치됨
ab{3}c : a와 c사이에 b가 3회 반복으로 나타나는 문자열과 매치됨
ab{1, 3}c : a와 c사이에 b가 1~3회 반복으로 나타나는 문자열과 매치됨
```

```python
1.1 : 1 + (한글로 된 모든 문자) + 1 인 문자열과 매치됨
<.*> : <>로 둘러싸인 모든 문자열과 매치됨
^\d.* : 숫자로 둘러싸인 모든 문자열과 매치됨
^\d.+\d$ : 숫자로 시작하고 숫자로 끝나는 문자열과 매치됨
\d{6}=\d{7} : 6개의 숫자-7개의 숫자 인 문자열과 매치됨
```
