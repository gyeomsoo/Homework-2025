# 132 클래스 변수와 인스턴스 변수

```python
class Wizard:
  generated = 0                             # 마법사 캐릭터 생성 수, 클래스 변수로 정의

  def __init__(self, name):                 # 클래스 생성자
    Wizard.generated += 1                   # 캐릭터가 생성되면 클래스 변수 1증가
    self.name = name                        # 캐릭터 이름 설정
    self.level = 1                          # 최초 레벨 1로 설정
    self.hp = 100                           # 최초 체력 10으로 설정
    self.initAbility()                      # 능력치 초기화

  def initAbility(self):                    # 능력치 초기화 메서드
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 10
    self.mentality = 10
    self.agility = 1
```

```python
player1 = Wizard('FirePlayer')
print(f'마법사생성 수: {player1.generated}\n캐릭터이름: {player1.name}')
```

```python
player2 = Wizard('GodGo')
print(f'마법사생성 수: {player2.generated}\n캐릭터이름: {player2.name}')
```

```python
player1.level = 2
print(f'player1.level={player1.level}, player2.level={player2.level}')
```

```python
player3 = Wizard('StormBaby')
print(player1.generated)
```

```python
class Wizard:
  generated = 0

  # 인스턴스 메서드
  def setFireballLevel(self, power=1):
    self.fireballLevel = power
    print(f'화염구 강도를 {power}로 설정합니다.')

  # 클래스 메서드
  @classmethod
  def addGenerated(cls, count):
    cls.generated += count
    print(f'생성한 총 캐릭터 수: {cls.generated}')

  # 정적 메서드
  @staticmethod
  def static1():
    print('Wizard 클래스에서 정의된 정적 메서드입니다.')
```

```python
player1 = Wizard()
player1.setFireballLevel(8)
Wizard.addGenerated(2)
Wizard.static1()
```
