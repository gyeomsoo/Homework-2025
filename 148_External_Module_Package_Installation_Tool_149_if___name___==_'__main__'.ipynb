# 149 if __ name __ == '__ main __' :

```python
def printName():
  print(f'module_b의 __name__ = {__name__}')
```

```python
from module_b import printName

printName()
print(f'실행 모듈인 이 모듈의 __name__ = {__name__}')
```

```python
def printName():
  print(f'module_b의 __name__ = {__name__}')

if __name__ == '__main__':
  print('이 모듈은 처음 실행되는 코드입니다.')
  printName()
```
