#136 클래스 상속

```python
class 자식클래스(부모클래스)
```

```python
class Character:
  def __init__(self):                # 생성자에서 캐릭터의 공통 능력치를 1로 초기
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 1
    self.mentality = 1
    self.agility = 1
```

```python
class Wizard(Character):
  def setBasic(self, name=None, level=1, hp=100):
    self.name = name
    self.offense_power = 1
    self.level = level
    self.hp = hp
```

```python
player1 = Wizard()
player1.setBasic('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

```python
class Wizard(Character):
  def __init__(self, name=None):
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

```python
class Wizard(Character):
  def __init__(self, name=None):
    super().__init__()
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
player1 = Wizard('StormBaby')
print(f'[{player1.name}] 레벨: {player1.level}')
print(f'정신력: {player1.mentality}')
```

# 137 예외처리 1 : try ~ except

```python
fileList = ['file01.jpg', 'file02.jpg', 'file03', 'file04.jpg']
for file in fileList:
  fname = file.split('.')
  print(f'파일 이름: {fname[0]}, 파일 확장자: {fname[1]}')
```

```python
for file in fileList:
  try:
    fname = file.split('.')
    print(f'파일 이름: {fname[0]}, 파일 확장자: {fname[1]}')
  except:
    print(f'ERROR!! <{file}> 처리중 예외 발생!')
```

```python
try:
  잠재적으로 오류 및 예외가 발생할 수 있는 코드
except:
  예외가 발생했을 때 예외를 처리하는 코드
```
