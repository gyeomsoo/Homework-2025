# 112 딕셔너리에 특정 키가 없는지 확인 : *key* not in *d*

```python
frequency = {}
while True:
  c = input('알파벳 1자를 입력하세요[끝내려면 엔터를 치세요]:')
  if c == '':
    break
  if c not in frequency:
    frequency[c] = 1
  else:
    frequency[c] += 1
print(frequency)
```

# 113 닥셔너리의 모든 키로 이터레이터 생성 : iter (*d*)

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
for name in iter(greats):
  print(name)
```
