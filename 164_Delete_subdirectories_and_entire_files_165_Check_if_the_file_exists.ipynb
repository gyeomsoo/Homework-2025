# 164 하위 디렉토리 및 파일 전체 삭제 : shutil.rmtree ()

```python
from shutil import rmtree

folder = 'c:/mywork/jupyter/myfolder'
k = input(f'<{folder}>를 완전히 삭제하시겠습니까? (y/n)')
if k lower() == 'y':
  rmtree(folder)
  print(f'<{folder}>를 완전히 삭제했습니다.')
else:
  print(f'<{folder}>를 삭제하지 않습니다.')
```

# 165 파일이 존재하는지 체크 : os.path.exists ()

```python
from os.path import exists

print(exists('sunrise.jpg'))
print(exists('tmp'))
print(exists('C:/mywork/jupyter/bts.mp4'))
```

```python
from os import mkdir
from os.path import exists

while True:
  folder = input('생성할 디렉토리 이름을 입력하세요: ')
  if exists(folder):
    print(f'<{folder}>가 이미 존재합니다.')
    continue
  else:
    mkdir(folder)
    print(f'<{folder}>를 생성했습니다.')
    break
```
