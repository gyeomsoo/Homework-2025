# 175 에코 서버 만들기 : socket

```python
from typing import dataclass_transform
import socket
HOST = ''     # 빈 문자열이면 서버가 구동되는 컴퓨터의 IP를 자동 할당
PORT = 7777

def echoServer():
  with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as sock:
    sock.bind((HOST, PORT))
    sock.listen()
    print('클라이언트 연결을 기다리는 중...')
    conn, addr = sock.accept()
    with conn:
      print(f'{addr[0]}와 연결됨')
      while True:
        data = conn.recv(1024)
        if not data:
          break
          print(f'수신 메시지: {data.decode()}')
          conn.sendall(data)

echoServer()
```
