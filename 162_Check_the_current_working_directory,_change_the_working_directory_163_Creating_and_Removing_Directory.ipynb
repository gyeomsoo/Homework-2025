# 162 현재 작업 디렉토리 확인, 작업 디렉토리 변경 : os.getcwd (), os.chdir ()

```python
from os import getcwd

print(f'현재 작업 디렉토리: {getcwd()}')
```

```python
from os import chdir

print(f'현재 작업 디렉토리: {getcwd()}')
chdir('..')                                 # 상대경로 지정
print(f'현재 작업 디렉토리: {getcwd()}')
chdir('C:/mywork/jupyter')                  # 절대경로 지정
print(f'현재 작업 디렉토리: {getcwd()}')
```

# 163 디렉토리 생성 및 제거 : os.mkdir (), os.rmdir ()

```python
from os import mkdir

folder = 'myfolder'
mkdir(folder)
```

```python
folder = 'C:/mywork/jupyter/myfolder2'
mkdir(folder)
```

```python
from os import rmdir

folder = 'myfolder'
forder2 = 'C:/mywork/jupyter/myfolder2'
rmdir(folder)
rmdir(forder2)
```
