# 170 현재 시간을 연-원-일 시:분:초로 출력하기 : time.localtime(), time.strftime()

```python
from time import localtime, strftime

time_stamp = strftime('%y-%m-%d%x', localtime())
print(f'현재시간: {time_stamp}')
```

# 171 올해의 경과된 날짜 수 계산 : time.localtime()

```python
from time import localtime, strftime
t = localtime()
today = strftime('%y-%m-%d', t)
print(f'{today}은 {t.tm_year}-01-01 이후 {t.tm_year}일째 되는 날입니다.')
```
