# 168 프로그램 실행 소요 시간 계산 : time.time ()

```python
from time import time
print(f'1970-01-01 00:00:00 이후 지금 현재까지 경과된 초: {time()}')
```

```python
def cumalativeSum(n):
  ret = 0
  for i in range(1, n+1):
    ret += i
  return ret
```

```python
from time import time
n = 10000000
stime = time()
answer = cumalativeSum(n)
etime = time()
print(f'1에서 {n:,}까지 누적합 계산 시간: {(etime-stime):.3f}초')
```

# 169 주어진 시간동안 일시정지 : time.sleep ()

```python
from time import sleep

for i in range(1, 6):
  print('*'*i)
  sleep(0.5)
```
