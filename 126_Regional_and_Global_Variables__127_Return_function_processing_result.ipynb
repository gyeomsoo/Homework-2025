# 127 함수 처리 결과 리턴 : return

```python
def reverse(*args):
  result = []
  idx = len(args) - 1
  while idx >= 0:
    result.append(args[idx])
    idx -= 1
  return result
```

```python
ret = reverse('가', '나', '다', '라')
print(ret)
```
